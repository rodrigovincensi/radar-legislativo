-- Extração dos Dados --

```python
import requests
import json
```

Extração Dados - Proposições

```python
url = "https://dadosabertos.camara.leg.br/api/v2/proposicoes"
params = {'itens': 100}
todas_proposicoes = []
max_retries = 5
retry_count = 0
```

```python
while url:
    try:
        response = requests.get(url, params=params)
        dados = response.json()
        todas_proposicoes.extend(dados['dados'])
        
        next_link = None
        for link in dados['links']:
            if link['rel'] == 'next':
                next_link = link['href']
        
        url = next_link
        retry_count = 0
    except:
        retry_count += 1
        if retry_count > max_retries:
            print(f"Erro na URL: {url}")
            break
```

```python
with open(r'C:\Rodrigo\Xperiun\Pós Tech\Squad\Projeto Integrador I\Arquivos Json\proposicoes_raw.json', 'w', encoding='utf-8') as f:
    json.dump(todas_proposicoes, f, ensure_ascii=False)
```

Extração Dados - Despesas

```python
with open(r'C:\Rodrigo\Xperiun\Pós Tech\Squad\Projeto Integrador I\Arquivos Json\deputados_raw.json', 'r', encoding='utf-8') as arquivo:
    dados = json.load(arquivo)
```

```python
lista_id_deputados = []
for deputado in dados:
    lista_id_deputados.append(deputado['id'])

todas_despesas = []
max_retries = 5
retry_count = 0
params = {'itens': 100}
```

```python
erro_fatal = False

for id_deputado in lista_id_deputados:
    url = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/despesas"

    while url:
        try:
            response = requests.get(url, params=params)
            dados = response.json()
            
            for despesa in dados['dados']:
                despesa['id_deputado'] = id_deputado
            
            todas_despesas.extend(dados['dados'])

            next_link = None
            for link in dados['links']:
                if link['rel'] == 'next':
                    next_link = link['href']
            
            url = next_link
            retry_count = 0
        except:
            retry_count += 1
            if retry_count > max_retries:
                erro_fatal = True
                break

    if erro_fatal:
        print(f"Erro fatal no deputado {id_deputado}. Extração interrompida.")
        break
```

```python
with open(r'C:\Rodrigo\Xperiun\Pós Tech\Squad\Projeto Integrador I\Arquivos Json\despesas_raw.json', 'w', encoding='utf-8') as f:
    json.dump(todas_despesas, f, ensure_ascii=False)
```
