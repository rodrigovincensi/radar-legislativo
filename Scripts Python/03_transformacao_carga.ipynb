-- Transformação e Carga --

```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv(r'C:\Rodrigo\Xperiun\Pós Tech\Squad\Projeto Integrador I\.env')
engine = create_engine(os.getenv('DATABASE_URL'))
print('Conexão configurada.')
```

Carregamento dos JSONs

```python
pasta = r'C:\Rodrigo\Xperiun\Pós Tech\Squad\Projeto Integrador I\Arquivos Json'

df_deputados   = pd.read_json(pasta + '\\deputados_raw.json')
df_partidos    = pd.read_json(pasta + '\\partidos_raw.json')
df_proposicoes = pd.read_json(pasta + '\\proposicoes_raw.json')
df_votacoes    = pd.read_json(pasta + '\\votacoes_raw.json')
df_despesas    = pd.read_json(pasta + '\\despesas_raw.json')

print('JSONs carregados.')
```

Transformação — Deputados

```python
df_deputados = df_deputados[['id','nome','siglaPartido','siglaUf','idLegislatura','urlFoto','email']]
df_deputados = df_deputados.drop_duplicates(subset='id')
print(f'Deputados: {len(df_deputados)} registros')
df_deputados.head()
```

Transformação — Partidos

```python
df_partidos = df_partidos[['id','sigla','nome']]
df_partidos = df_partidos.drop_duplicates(subset='id')
print(f'Partidos: {len(df_partidos)} registros')
df_partidos.head()
```

Transformação — Proposições

```python
df_proposicoes = df_proposicoes[['id','siglaTipo','codTipo','numero','ano','ementa','dataApresentacao']]
df_proposicoes['dataApresentacao'] = pd.to_datetime(df_proposicoes['dataApresentacao'], errors='coerce')
df_proposicoes = df_proposicoes.drop_duplicates(subset='id')
print(f'Proposicoes: {len(df_proposicoes)} registros')
df_proposicoes.head()
```

Transformação — Votações

```python
df_votacoes = df_votacoes[['id','data','dataHoraRegistro','siglaOrgao','proposicaoObjeto','descricao','aprovacao']]
df_votacoes['data'] = pd.to_datetime(df_votacoes['data'], errors='coerce')
df_votacoes['dataHoraRegistro'] = pd.to_datetime(df_votacoes['dataHoraRegistro'], errors='coerce')
df_votacoes = df_votacoes.drop_duplicates(subset='id')
print(f'Votacoes: {len(df_votacoes)} registros')
df_votacoes.head()
```

Transformação — Despesas

```python
df_despesas = df_despesas[['id_deputado','ano','mes','tipoDespesa','codDocumento','tipoDocumento','dataDocumento','numDocumento','valorDocumento','nomeFornecedor','cnpjCpfFornecedor','valorLiquido','valorGlosa']]
df_despesas['dataDocumento'] = pd.to_datetime(df_despesas['dataDocumento'], errors='coerce')
df_despesas = df_despesas[df_despesas['valorDocumento'] >= 0]
df_despesas = df_despesas.drop_duplicates()
print(f'Despesas: {len(df_despesas)} registros')
df_despesas.head()
```

Carga no Supabase

```python
from sqlalchemy import inspect

# Deputados, partidos, votacoes, despesas: replace seguro (sem dados de IA)
for nome, df in [('deputados', df_deputados), ('partidos', df_partidos), ('votacoes', df_votacoes), ('despesas', df_despesas)]:
    df.to_sql(nome, engine, if_exists='replace', index=False)
    print(f'{nome}: {len(df)} registros carregados')

# Proposicoes: insert-only para preservar a coluna resumo gerada pelo notebook 04
tabela_existe = inspect(engine).has_table('proposicoes')

if not tabela_existe:
    df_proposicoes.to_sql('proposicoes', engine, if_exists='replace', index=False)
    print(f'proposicoes: {len(df_proposicoes)} registros carregados (primeira carga)')
else:
    with engine.connect() as conn:
        ids_existentes = set(pd.read_sql('SELECT id FROM proposicoes', conn)['id'].tolist())
    df_novas = df_proposicoes[~df_proposicoes['id'].isin(ids_existentes)]
    if len(df_novas) > 0:
        df_novas.to_sql('proposicoes', engine, if_exists='append', index=False)
    print(f'proposicoes: {len(df_novas)} novas inseridas, {len(ids_existentes)} existentes preservadas')

print('Carga concluída.')
```
